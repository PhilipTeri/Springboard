```python
import sqlite3
from sqlite3 import Error

 
def create_connection(db_file):
    """ create a database connection to the SQLite database
        specified by the db_file
    :param db_file: database file
    :return: Connection object or None
    """
    conn = None
    try:
        conn = sqlite3.connect(db_file)
        print(sqlite3.version)
    except Error as e:
        print(e)
 
    return conn

 
def select_all_tasks(conn):
    """
    Query all rows in the tasks table
    :param conn: the Connection object
    :return:
    """
    cur = conn.cursor()
    
   

    # Q10: Produce a list of facilities with a total revenue less than 1000.
    #The output of facility name and total revenue, sorted by revenue. Remember
    #that there's a different cost for guests and members!
    
    query10 = """
        SELECT facility_name, SUM(Revenue)

        FROM 

            (SELECT facility.name AS facility_name,

             CASE

                WHEN bookings.memid = 0 THEN facility.guestcost * bookings.slots
                WHEN bookings.memid > 0 THEN facility.membercost * bookings.slots

                END AS Revenue

             FROM `Bookings` AS bookings


             LEFT JOIN `Facilities` AS facility

             USING (facid)) AS sub


        GROUP BY facility_name HAVING SUM(revenue) < 1000

        ORDER BY SUM(Revenue) DESC

        """
    
  
    #Q11: Produce a report of members and who recommended them in alphabetic surname,firstname order
    
    query11 = """
    SELECT members.memid, members.firstname || ' ' ||  members.surname AS member_name, members.recommendedby AS recommended, recommenders.firstname || ' ' ||  recommenders.surname AS recommenders_name 

    FROM `Members` AS members


    INNER JOIN `Members` AS recommenders


    ON members.recommendedby = recommenders.memid


    WHERE members.memid > 0 AND members.recommendedby != ''


    ORDER BY member_name
    
    """
    
    

    #/* Q12: Find the facilities with their usage by member, but not guests */
    
    query12 = """
    
    SELECT facilities.name, bookings.memid, COUNT(bookings.bookid), members.firstname || ' ' ||  members.surname AS member_name 


    FROM `Bookings` as bookings


    LEFT JOIN `Members` AS members

    USING(memid)

    LEFT JOIN `Facilities` AS facilities

    using(facid)


    WHERE bookings.memid != 0

    GROUP BY bookings.memid 


    """
   
    
    
    
    #/* Q13: Find the facilities usage by month, but not guests */
    
    query13 = """
    SELECT bookings.facid, facility.name, COUNT(bookings.bookid), strftime('%m', bookings.starttime) AS mon


    FROM `Bookings` AS bookings
    
    LEFT JOIN `Facilities` AS facility
    USING(facid)

    WHERE memid != 0
    
    GROUP by mon, bookings.facid

    """

    cur.execute(query12)
   
 
    rows = cur.fetchall()
 
    for row in rows:
        print(row)


def main():
    database = "sqlite_db_pythonsqlite.db"
 
    # create a database connection
    conn = create_connection(database)
    with conn: 
        print("2. Query all tasks")
        select_all_tasks(conn)
 
 
if __name__ == '__main__':
    main()
```
